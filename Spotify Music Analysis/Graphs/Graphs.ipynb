```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")

pd.options.display.max_rows= None
```

```python
months=pd.read_csv('monthsgenres.csv')
week=pd.read_csv('weekgenres.csv')
genr2020=pd.read_csv('genr2020.csv')
genr2019=pd.read_csv('genr2019.csv')
genr2018=pd.read_csv('genr2018.csv')
genr2017=pd.read_csv('genr2018.csv')
```

```python
#Last week Genres

plt.figure(figsize=(11,6))
sns.color_palette("light:#5A9", as_cmap=True)
wk=sns.barplot(x="genre", y="factor", data=week, palette=["C3", "C1", "C2", 'C10', "C4", 'C7'])
wk.set(xlabel='Genres', ylabel='Popularity', title='Top Music Genres- Week 4 of 2021')
# plt.xticks(rotation=45)

plt.show()
```
