```python
import pandas as pd

data=pd.read_csv('cleaned_data.csv')

pd.options.display.max_columns = None
pd.options.display.max_columns = None
data
```

```python
data.sort_values(['end_week'], ascending= False)
```

```python
data['factor']=1/data['Position']

data
```

```python
data.drop_duplicates(inplace= True)

data
```

```python
today=data[data['end_week']=='2021-01-22']

today
```

```python
today.drop_duplicates(inplace= True)

today
```

```python
week=today.groupby(['genre'], as_index= False)['factor'].sum()

week.sort_values(['factor'], ascending= False, inplace=True, ignore_index=True)

week
```

```python
lastmonths=data[data['end_week']>'2020-08-28']
```

```python
months=lastmonths.groupby(['genre'], as_index= False)['Streams'].sum()

months.sort_values(['Streams'], ascending= False, inplace=True, ignore_index=True)

months
```

```python
#data for 2021

d2021=data[data['end_week']>'2020-12-25']
d2021
```

```python
art2021=d2021.groupby(['artist_name'], as_index= False)['Streams'].sum()

art2021.sort_values(['Streams'], ascending= False, inplace=True, ignore_index=True)

t2021=art2021.head(100)

t2021
```

```python
daily=pd.read_csv('regional-global-daily-2021-05-20.csv')

daily
```

```python
daily.rename(columns = {'Unnamed: 0':'Position','Unnamed: 1':'Track Name','Unnamed: 2':'Artist','Note that these figures are generated using a formula that protects against any artificial inflation of chart positions.':'Streams','Unnamed: 4':'URL'}, inplace = True)
daily.drop(daily.index[0], inplace=True)

daily.drop_duplicates(subset='Artist', keep='first' , inplace=True)

daily=daily.head(100)
daily
```

```python
top100=daily[['Artist']].head(100)

len(top100['Artist'].unique())
```

```python
top2021=t2021[(t2021['artist_name'].isin(top100['Artist']))]

top2021.reset_index(inplace=True, drop=True)

del top2021['Streams']
top2021
```

```python
top2020=t2020[(t2020['artist_name'].isin(top100['Artist']))]

top2020.reset_index(inplace=True, drop=True)
del top2020['Streams']
top2020
```

```python
top2019=t2019[(t2019['artist_name'].isin(top100['Artist']))]

top2019.reset_index(inplace=True, drop=True)
del top2019['Streams']
top2019
```

```python
top2018=t2018[(t2018['artist_name'].isin(top100['Artist']))]

top2018.reset_index(inplace=True, drop=True)

del top2018['Streams']
top2018
```

```python
top2017=t2017[(t2017['artist_name'].isin(top100['Artist']))]

top2017.reset_index(inplace=True, drop=True)

del top2017['Streams']

top2017
```

```python
every=t2018[(t2018['artist_name'].isin(top100['Artist']))]
every=every[(every['artist_name'].isin(t2018['artist_name']))]
every=every[(every['artist_name'].isin(t2019['artist_name']))]
every=every[(every['artist_name'].isin(t2020['artist_name']))]

every.reset_index(inplace=True, drop=True)

del every['Streams']

every
```
