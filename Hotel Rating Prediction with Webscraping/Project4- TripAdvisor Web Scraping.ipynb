```python
name=[]
link=[]
id=[]
sponsored=[]
provider=[]
price=[]
rating=[]
reviews=[]
free_wifi=[]
restaurant=[]
covid_safe=[]

for p in tqdm(page):
    main_link='https://www.tripadvisor.com.gr/Hotels-g298566-{}-Osaka_Osaka_Prefecture_Kinki-Hotels.html'.format(p)
    response = requests.get(main_link,timeout=30)#, headers= user_agent)
    soup = BeautifulSoup(response.content, 'html.parser')
    msoup=soup.find_all('div', class_='ui_column is-8 main_col allowEllipsis')
    for i in range (0,len(msoup)):
        #name
        name.append(msoup[i].find_all('div', class_='prw_rup prw_meta_hsx_listing_name listing-title')[0]('a')[0].text.strip())
        #link
        a=msoup[i].find_all('div', class_='prw_rup prw_meta_hsx_listing_name listing-title')[0]('a')[0].get('href')
        link.append('https://www.tripadvisor.com.gr/'+a)
        #id
        id.append(msoup[i].find_all('a', class_='property_title prominent')[0].get('id'))
        #is sponsored
        try:    
            a=msoup[i].find_all('div', class_='ui_column is-narrow')[0].text
            sponsored.append(1)
        except:
            sponsored.append(0)
        #provider
        try:
            provider.append(msoup[i].find_all('div', class_="premium_offer no_cpu ui_columns is-mobile is-gapless is-multiline with-condensed-primary-commerce with-provider-logo metaOffer")[0].find_all('span', class_='provider_text')[0].text )
        except:
            provider.append(np.nan)
        #price
        try:
            price.append(msoup[i].find_all('div', class_="text-link no_cpu metaOffer" )[0].get('data-pernight'))
        except:
            price.append(np.nan)
        #rating
        try:
            r=msoup[i].find_all('div', class_="prw_rup prw_common_rating_and_review_count_with_popup linespace is-shown-at-mobile")[0].find_all('a')[0].get('class')[1][-2:]
            rating.append(r[:1]+'.'+r[-1:])
        except:
            rating.append(np.nan)
        #reviews
        try:
            b=msoup[i].find_all('div', class_="prw_rup prw_common_rating_and_review_count_with_popup linespace is-shown-at-mobile")[0].find_all('a')[1].text
            num_of_reviews=re.sub("[^0-9]", "", b)
            reviews.append(num_of_reviews)
        except:
            reviews.append(np.nan)
        #free_wifi
        try:    
            a=msoup[i].find_all('span', class_='text')[0].text
            free_wifi.append(1)
        except:
            free_wifi.append(0)
        #restaurant
        try:    
            a=msoup[i].find_all('span', class_="text_container")[1].text
            restaurant.append(1)
        except:
            restaurant.append(0)
        #covid_safe
        try:    
            a=msoup[i].find_all('span', class_='_1shPZD63')[0].text
            covid_safe.append(1)
        except:
            covid_safe.append(0)
```

```python
df = pd.DataFrame(columns=['name', 'id', 'sponsored', 'provider','price','rating','reviews','free_wifi','restaurant','covid_safe', 'link'])

df['name']=name
df['link']=link
df['id']=id
df['sponsored']=sponsored
df['provider']=provider
df['price']=price
df['rating']=rating
df['reviews']=reviews
df['free_wifi']=free_wifi
df['covid_safe']=covid_safe
df['restaurant']=restaurant

df
```

```python
df.to_csv('Osaka_Stays.csv', index=False)
```
